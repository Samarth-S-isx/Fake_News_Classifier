```python
df=df.dropna()
df = df.reset_index()
X = df.drop('label',axis=1)
y = df['label']
```

```python
messages = df['title']
messages.size
```

```python
y =df['label']
```

```python
import re
corpus1 = []
for i in range(len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages[i])
    review = review.lower()
    review = review.split()
    review = ' '.join(review)
    corpus1.append(review)
len(corpus1)
```

```python
import spacy
nlp = spacy.load("en_core_web_lg")
```

```python
nlp.meta['vectors']
```

```python
corpus = []
for d in corpus1:
    doc1 = nlp(d)
    word = [w.lemma_ for w in doc1 if nlp.vocab[str(w)].is_stop==False]
    corpus.append(nlp(' '.join(word)))
corpus
```
